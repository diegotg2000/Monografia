```python
%pylab inline
import pandas as pd
from astropy.cosmology import WMAP9 as cosmo
```

```python
data=np.loadtxt('BigMD_data.dat')
```

```python
dec=data[:,1]*np.pi/180
```

```python
ra=data[:,0]*np.pi/180
```

```python
plt.scatter(ra,dec,s=0.001)
```

```python
redshift=data[:,2]
```

```python
r=np.array(cosmo.comoving_distance(redshift))
```

```python
x=r*np.cos(ra)*np.cos(dec)
y=r*np.sin(ra)*np.cos(dec)
z=r*np.sin(dec)
```

```python
A=np.zeros((len(x),3))
```

```python
A[:,0]=x
A[:,1]=y
A[:,2]=z
```

```python
A[:5]
```

```python
np.savetxt('bMD.txt',A)
```
