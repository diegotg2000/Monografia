```python
%pylab inline
import numpy as np
```

```python
x=np.arange(1.0,5.1,0.5)
```

```python
plt.figure(figsize=(7,6))
for name,col,mark,label in zip(['obs','J08','Patchy','bMD','abacus'],['firebrick','midnightblue','black','darkgreen','purple'],['o','*','x','+','^']
                              ,['BOSS data','J08','Patchy','BigMD','Abacus']):
    y=np.loadtxt('C_'+name+'.txt')
    plt.plot(x,y,label=label,color=col,marker=mark,alpha=0.8)
plt.legend(frameon=False,fontsize=12)
plt.xlabel('\u03b2',fontsize=18)
plt.ylabel('C',fontsize=18)
plt.xticks(np.arange(1.0,5.1,0.5),fontsize=12)
plt.yticks(np.arange(0.0,0.14,0.02),fontsize=12)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.savefig('comparacion.pdf')
```

```python
Y=np.loadtxt('C_obs.txt')
```

```python
plt.figure(figsize=(7,6))
for name,col,mark,label in zip(['obs','J08','Patchy','bMD','abacus'],['firebrick','midnightblue','black','darkgreen','purple'],['o','*','x','+','^']
                              ,['Observacional','J08','Patchy','BigMD','Abacus']):
    y=np.loadtxt('C_'+name+'.txt')-Y
    plt.plot(x,y,label=label,color=col,marker=mark,alpha=0.8)
plt.legend(frameon=False,fontsize=12)
plt.xlabel('\u03b2',fontsize=18)
plt.ylabel('$C-C_{obs}$',fontsize=18)
plt.xticks(np.arange(1.0,5.1,0.5),fontsize=12)
plt.yticks(np.arange(0.0,0.041,0.005),fontsize=12)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
```
