```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy as sp
import pandas as pd
from scipy import stats
```

```python
#Análisis inicial
#Entropía
x=np.loadtxt('Betas_mas_puntos.txt')
y1=np.loadtxt('S_norandom_mas_puntos.txt')
y2=np.loadtxt('S_random_mas_puntos.txt')
plt.figure(figsize=(5,4))
plt.plot(x,y1,label='Simulation',c='black',ls='-')
plt.plot(x,y2,label='Random',c='blue',ls='--')
plt.xlabel('\u03b2',fontsize=18)
plt.ylabel('S',fontsize=18)
plt.legend(fontsize=12)
plt.xticks([1,2,3,4,5],fontsize=12)
plt.yticks(np.arange(16,33,4)/10,fontsize=12)
plt.tight_layout()
plt.savefig('svb.pdf')
```

```python
#Varianza
x=np.loadtxt('Betas_varianza.txt')
plt.figure(figsize=(5,4))
for i in range(1,21):
    y=np.loadtxt('CvB_{}.txt'.format(i))
    plt.plot(x,y,c='black')
plt.xlabel('\u03b2',fontsize=12)
plt.ylabel('C',fontsize=12)
```

```python
#VarianzaNueva
plt.figure(figsize=(5,4))
x=np.loadtxt('Betas_varianza.txt')
A=np.loadtxt('CvB_1.txt')
for i in range(1,21):
    y=np.loadtxt('CvB_{}.txt'.format(i))-A
    plt.plot(x,y,c='black',linewidth=0.7)
plt.xlabel('\u03b2',fontsize=18)
plt.ylabel('C - C$_0$',fontsize=18)
plt.text(4.9,10.5/1000,'a)',fontsize=18)
plt.yticks(np.arange(-4,13,4)/1000,fontsize=12)
plt.xticks(fontsize=12)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.tight_layout()
plt.savefig('varianza.pdf')
```

```python
#VarianzaNueva2
plt.figure(figsize=(5,4))
x=np.loadtxt('Betas_varianza.txt')
ymean=np.zeros(len(x))
for i in range(1,21):
    y=np.loadtxt('CvB_{}.txt'.format(i))
    ymean+=y
ymean=ymean/20
ysigma=np.zeros(len(x))
for i in range(1,21):
    y=np.loadtxt('CvB_{}.txt'.format(i))
    dy2=(y-ymean)**2
    ysigma+=dy2
ysigma=np.sqrt(ysigma/20)
plt.plot(x,ysigma,label='sigma')
sigmaSpheres=np.loadtxt('sigmacomp_spheres.txt')
plt.plot(x,sigmaSpheres/ymean,label='sigma/mean')
plt.xlabel('\u03b2',fontsize=18)
plt.ylabel('(C - C$_0$)/C$_0$',fontsize=18)
plt.legend()
plt.savefig('varianza_Test.pdf')
```

```python
#Geometria
plt.figure(figsize=(5,4))
x=np.loadtxt('Betas_geom.txt')
spheres=np.loadtxt('meancomp_spheres.txt')
shells=np.loadtxt('meancomp_shells.txt')
sigmaSpheres=np.loadtxt('sigmacomp_spheres.txt')
sigmaShells=np.loadtxt('sigmacomp_shells.txt')
plt.errorbar(x,spheres,sigmaSpheres, label='Spheres', c='black')
plt.errorbar(x,shells,sigmaShells, label='Shells', c='blue',ls='--')
plt.xlabel('\u03b2',fontsize=18)
plt.ylabel('C',fontsize=18)
plt.legend(loc=3,fontsize=12)
plt.text(4.9,12.8/100,'b)',fontsize=18)
plt.yticks(np.arange(0,15,2)/100,fontsize=12)
plt.xticks(fontsize=12)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.tight_layout()
plt.savefig('geometria.pdf')
```

```python
#RSD
plt.figure(figsize=(5,4))
x=np.loadtxt('Betas_RSD.txt')
RSD=np.loadtxt('meancomp_RSD.txt')
NoRSD=np.loadtxt('meancomp_NoRSD.txt')
sigmaRSD=np.loadtxt('sigmacomp_RSD.txt')
sigmaNoRSD=np.loadtxt('sigmacomp_NoRSD.txt')
plt.errorbar(x,NoRSD,sigmaNoRSD, label='No RSD', c='black')
plt.errorbar(x,RSD,sigmaRSD, label='RSD', c='blue',ls='--')
plt.xlabel('\u03b2',fontsize=18)
plt.ylabel('C',fontsize=18)
plt.text(4.9,14.5/100,'c)',fontsize=18)
plt.legend(loc=3,fontsize=12)
plt.yticks(np.arange(0,17,4)/100,fontsize=12)
plt.xticks(fontsize=12)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.tight_layout()
plt.savefig('rsd.pdf')
```

```python
#RedshiftEvolution
plt.figure(figsize=(5,4))
x=np.loadtxt('zetas.txt')
marcadores=['o','P','^','s','D']
colores=['darkcyan','black','red','darkgreen','blue']
for i,j in enumerate([1.0,1.5,2.0,2.5,3.0]):
    y=np.loadtxt('beta={}.txt'.format(j))
    color=colores[i]
    plt.plot(x,y,label='\u03b2={}'.format(j),marker=marcadores[i],c=color)
plt.xlabel('z',fontsize=18)
plt.ylabel('C - C$_{0.1}$',fontsize=18)
plt.xticks(x,fontsize=12)
plt.legend(fontsize=12)
plt.text(1.45,10.4/1000,'d)',fontsize=18)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.yticks(np.arange(-4,13,4)/1000,fontsize=12)
plt.xticks(fontsize=12)
plt.tight_layout()
plt.savefig('rsd_evolution.pdf')
```

```python
#Clustering
x=np.loadtxt('Betas_porcen.txt')
obs=np.loadtxt('C_obs.txt')
plt.figure(figsize=(5,4))
for i in range(10,110,10):
    y=np.loadtxt('p_{}.txt'.format(i))
    width=2*i/100+0.5
    if i in [10,30,50,70,100]:
        plt.plot(x,y,label='{}%'.format(i),c='black',lw=width)
    else:
        plt.plot(x,y,c='black',lw=width)
#plt.plot(x,obs,label='Observational',c='red')
plt.xlabel('\u03b2',fontsize=18)
plt.ylabel('C',fontsize=18)
plt.legend(fontsize=12)
plt.yticks(np.arange(0,15,2)/100,fontsize=12)
plt.xticks(np.arange(1,6),fontsize=12)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.tight_layout()
plt.savefig('cvb_porc.pdf')
```
