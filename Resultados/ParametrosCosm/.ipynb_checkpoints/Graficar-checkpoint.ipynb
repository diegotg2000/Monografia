```python
for number in ['1.0','2.0','3.5','5.0']:
    plt.figure(figsize=(10,15),tight_layout=False)
    plt.suptitle('\u03b2 = '+number,x=0.5,y=0.93,fontsize=15)
    for i in range(len(PARAMETROS)):
        plt.subplot(3,2,i+1)
        x=np.loadtxt('box_cosmo_params.dat',usecols=i+1)
        y=np.loadtxt('beta='+number,usecols=0)
        rho=sp.stats.spearmanr(y,x)[0]
        plt.scatter(x,y,s=7,c='darkblue)
        plt.xlabel(PARAMETROS[i],fontsize=13)
        plt.ylim((min(y)-std(y)/3,max(y)+std(y)/3))
        if i==2:
            plt.ylabel('C',fontsize=13)
        plt.legend()
    plt.savefig('CvB_param_'+number+'.pdf')
```

```python
plt.figure(figsize=(5,4))
BETAS=[1.0,2.0,3.5,5.0]
marcadores=['o','X','^','s','D','p']
colores=['darkcyan','black','red','darkgreen','blue', 'purple']
for j in range(len(PARAMETROS)):
    x=np.loadtxt('box_cosmo_params.dat',usecols=j+1)
    Y=np.zeros(len(BETAS))
    for i in range(len(BETAS)):
        y=np.loadtxt('beta='+str(BETAS[i]),usecols=0)
        rho=sp.stats.spearmanr(y,x)[0]
        Y[i]=rho
    plt.plot(BETAS, Y, label=str(PARAMETROS[j]),marker=marcadores[j],c=colores[j])
plt.xlabel('\u03b2',fontsize=18)
plt.ylabel('\u03c1',fontsize=18)
plt.xticks(BETAS,fontsize=12)
plt.yticks(np.arange(-5,6,2)/10,fontsize=12)
plt.legend(fontsize=12)
plt.tight_layout()
plt.savefig('rhovb.pdf')
```

```python
x=np.loadtxt('box_cosmo_params.dat',usecols=5)
y=np.loadtxt('beta=5.0')
rho=sp.stats.spearmanr(y,x)[0]
plt.figure(figsize=(5,4))
plt.scatter(x,y,label='\u03c1 = {}'.format(np.round(rho,3)),c='blue',s=7)
plt.ylim((min(y)-std(y)/3,max(y)+std(y)/3))
plt.xlabel('$\sigma_{8}$',fontsize=18)
plt.ylabel('C',fontsize=18)
plt.title('\u03b2=5.0',fontsize=18)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.xticks(np.arange(62,105,6)/100,fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.tight_layout()
plt.savefig('cvsigma8.pdf')
```
